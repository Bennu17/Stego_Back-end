```python
import matplotlib.pyplot as plt

import numpy as np

from PIL import Image



def plot_rgb_intensity(image_path):

  # Load image

  image = Image.open(image_path)

  image = image.convert('RGB') # Ensure image is in RGB mode



  # Convert image to numpy array

  image_array = np.array(image)



  # Extract RGB channels

  red_channel = image_array[:, :, 0].flatten()

  green_channel = image_array[:, :, 1].flatten()

  blue_channel = image_array[:, :, 2].flatten()



  # Calculate histogram for each channel

  bins = np.linspace(0, 256, 257)

  red_hist, _ = np.histogram(red_channel, bins=bins)

  green_hist, _ = np.histogram(green_channel, bins=bins)

  blue_hist, _ = np.histogram(blue_channel, bins=bins)



  # Plot the histograms

  plt.figure(figsize=(10, 5))

  plt.plot(bins[:-1], red_hist, color='red', alpha=0.6, label='Red Channel')

  plt.plot(bins[:-1], green_hist, color='green', alpha=0.6, label='Green Channel')

  plt.plot(bins[:-1], blue_hist, color='blue', alpha=0.6, label='Blue Channel')



  plt.xlabel('Intensity Value')

  plt.ylabel('Pixel Count')

  plt.title('RGB Intensity Distribution')

  plt.legend()

  plt.grid(True)

  plt.show()



# Example usage

image_path = 'C:\\Users\\Administrator\\Desktop\\testing\\sai_original1.png' # Replace with your image path

plot_rgb_intensity(image_path)

```

```python
import matplotlib.pyplot as plt

from PIL import Image



def plot_pixel_histogram(image_path, title):

  # Open the image

  image = Image.open(image_path)

   

  # Convert the image to grayscale

  image = image.convert('L')

   

  # Flatten the pixel values

  pixels = list(image.getdata())

   

  # Plot histogram

  plt.hist(pixels, bins=256, range=(0, 255), color='gray', alpha=0.7)

  plt.title(title)

  plt.xlabel('Pixel Value')

  plt.ylabel('Frequency')

  plt.show()



# Path to stego and non-stego images

stego_image_path = 'C:\\Users\\Administrator\\Desktop\\testing\\sai_original9.png'

non_stego_image_path = 'C:\\Users\\Administrator\\Desktop\\testing\\sai_original2.png'



# Plot histograms for stego and non-stego images

plot_pixel_histogram(stego_image_path, 'Stego Image Pixel Histogram')

plot_pixel_histogram(non_stego_image_path, 'Non-Stego Image Pixel Histogram')

```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the stego image and the original image
stego_image_path ='C:\\Users\\Administrator\\Desktop\\testing\\sai_original9.png'
original_image_path ='C:\\Users\\Administrator\\Desktop\\testing\\sai_original2.png'

stego_image = Image.open(stego_image_path)
original_image = Image.open(original_image_path)

# Convert images to numpy arrays
stego_array = np.array(stego_image)
original_array = np.array(original_image)

# Compute the difference between the two images
difference = np.abs(stego_array.astype(int) - original_array.astype(int))

# Plot the difference
plt.figure(figsize=(10, 5))

# Plot the original image
plt.subplot(1, 3, 1)
plt.imshow(original_array)
plt.title('Original Image')
plt.axis('off')

# Plot the stego image
plt.subplot(1, 3, 2)
plt.imshow(stego_array)
plt.title('Stego Image')
plt.axis('off')

# Plot the difference
plt.subplot(1, 3, 3)
plt.imshow(difference.astype(np.uint8))
plt.title('Difference')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("C:\\Users\\Administrator\\Desktop\\testing\\sai_original9.png", cv2.IMREAD_GRAYSCALE)

# Extract the LSB of each pixel
lsb_image = np.bitwise_and(image, 1)

# Plot the original and LSB images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# LSB image
plt.subplot(1, 2, 2)
plt.imshow(lsb_image, cmap='gray')
plt.title('LSB Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("C:\\Users\\Administrator\\Desktop\\testing\\sai_original3.png", cv2.IMREAD_GRAYSCALE)

# Extract the LSB of each pixel
lsb_image = np.bitwise_and(image, 1)

# Create a color mask for LSB values
color_mask = np.zeros_like(image, dtype=np.uint8)
color_mask[lsb_image == 0] = [255, 255, 255]  # White for LSB=0
color_mask[lsb_image == 1] = [255, 0, 0]      # Red for LSB=1

# Plot the original and colorized LSB images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Colorized LSB image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(color_mask, cv2.COLOR_BGR2RGB))
plt.title('Colorized LSB Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

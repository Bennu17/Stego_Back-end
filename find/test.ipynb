```python
import numpy as np

import os

from tensorflow.keras.applications import ResNet50, preprocess_input

from tensorflow.keras.preprocessing.image import load_img, img_to_array

from sklearn.cluster import KMeans

from sklearn.preprocessing import StandardScaler

from PIL import Image, UnidentifiedImageError



# Load the pre-trained ResNet50 model without the top layer (for feature extraction)

model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))



def extract_features(image_path):

  img = load_img(image_path, target_size=(224, 224))

  img_array = img_to_array(img)

  img_array = np.expand_dims(img_array, axis=0)

  img_array = preprocess_input(img_array)

   

  # Extract features using ResNet50

  features = model.predict(img_array)

  return features.flatten()



# Directory containing the images

folder_path = '"C:\Users\Administrator\Desktop\testing"'



# Get list of image file paths

image_paths = []

file_names = os.listdir(folder_path)



for file_name in file_names:

  if not file_name.startswith('.') and not file_name.endswith('.txt'):

    file_path = os.path.join(folder_path, file_name)

    try:

      img = Image.open(file_path)

      image_paths.append(file_path)

    except UnidentifiedImageError as e:

      print(f"Error processing image {file_path}: {e}")



# Extract features for each image

X = np.array([extract_features(image_path) for image_path in image_paths])



# Inspect the feature vectors

print("Feature vectors:\n", X)



# Normalize the features

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)





# Perform KMeans clustering

kmeans = KMeans(n_clusters=2, random_state=42)

labels = kmeans.fit_predict(X_scaled)



# Output the cluster labels for each image

for i, label in enumerate(labels):

  print(f'Image {image_paths[i]} is in cluster {label}')

```
